```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
stop_words
```

removestopwords

```python
def removestopwords(text):
    temp=[]
    for i in text:
        if i not in stop_words:
            temp.append(i)
    return temp
```

```python
t=t.apply(removestopwords)
t
```

stemminnng

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stemminnng(text):
    temp=[]
    for w in text:
       temp.append(ps.stem(w))
    return temp
```

```python
t=t.apply(stemminnng)
t
```
