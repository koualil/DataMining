```python
import matplotlib.pyplot as plt
import numpy as np

with open('ex1data1.txt', 'r') as file:
    data = file.read()
```

```python
x_values = []
y_values = []
lines = data.split('\n')
for line in lines:
    if line:  
        x, y = line.split(',')
        x_values.append(float(x))
        y_values.append(float(y))

# Création du diagramme de dispersion
plt.scatter(x_values, y_values, marker='+', color='red')

# Ajout d'un titre et des labels d'axes
plt.title("scater plot of training data")
plt.xlabel("population of city in 10000s")
plt.ylabel("profit in $10000s")

# Affichage du diagramme de dispersion
plt.show()
```

```python
def fit(x,y):
    x=np.array(x)
    y=np.array(y)
    alpha = 0.01
    theta0=0
    theta1=0
    m=len(y)
    for i in range(1000):
        py = theta0+theta1*x
        er=py-y
        squared_er = er**2
        MeanSqured_er = sum(squared_er)/len(y)
         
        dtheta1=-(1/m)*sum(x*(y-py))
        dtheta0=-(1/m)*sum(y-py)
        theta1=theta1-dtheta1*alpha
        theta0=theta0-dtheta0*alpha
    return theta0,theta1

theta0,theta1=fit(x_values , y_values)
theta0,theta1
```

```python
plt.scatter(x_values, y_values, marker='+', color='red')

# Ajout d'un titre et des labels d'axes
plt.title("scater plot of training data")
plt.xlabel("population of city in 10000s")
plt.ylabel("profit in $10000s")

 
plt.plot(x_values,theta1*np.array(x_values)+theta0 , color='red', linewidth=1)
# Affichage du diagramme de dispersion
# print("la valeur predite de 10.957 est "+str(a*10.957+o))
plt.show()
```
