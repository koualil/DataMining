```python
df['Text']= df['Text'].replace(to_replace=r"[^a-zA-Z ]+", value="")
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = stopwords.words('english')
```

# Tokenization

```python
def toknization(text):
    return text.split(" ")
```

```python
df['Text']=df['Text'].apply(toknization)
```

# Removal stopwords

```python
def removestopwords(text):
    temp=[]
    for i in text:
        if i not in stop_words:
            temp.append(i)
    return temp
```

```python
df['Text']=df['Text'].apply(removestopwords)
```

# Stemming

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stemminnng(text):
    temp=[]
    for w in text:
       temp.append(ps.stem(w))
    return temp
```

```python
df['Text']=df['Text'].apply(stemminnng)
df
```

# convert all texts in one big list

```python
corpus=df['Text'].tolist()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
preprocessed_data = [' '.join(tokens) for tokens in corpus]
```

```python
vectorizer = TfidfVectorizer()
```

```python
tfidf_matrix = vectorizer.fit_transform(preprocessed_data)
```

```python
tfidf_matrix
```

```python
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

df_tfidf = df_tfidf.T
```

```python
df_tfidf
```

```python
df_tfidf = df_tfidf.T
df_tfidf['Document_id'] = range(1, len(corpus) + 1)
df_tfidf.set_index('Document_id', inplace=True)

df_tfidf
```

```python
bow_matrix = df_tfidf.apply(lambda x: x * x.sum() / x[x != 0].sum(), axis=0)
# bow_matrix = bow_matrix.T
bow_matrix
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2,n_init=50, random_state=0)

kmeans.fit(bow_matrix)
```

```python
cluster_labels = kmeans.labels_
```

```python
cluster_labels
```

```python
df_predicted = pd.DataFrame(cluster_labels, columns=["target"])
```

```python
df['target']=df_predicted['target']
```

```python
df['target']=df_predicted['target']
df['IsToxic'] = df['IsToxic'].replace({False:0, True:1})
```
